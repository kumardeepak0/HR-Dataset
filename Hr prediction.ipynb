```python
def empirical_analysis(col,b):
    cold = []
    for i in col:
        if i == b:
            cold.append(i)
    
    col = cold[0]
    mu = hrdf[col].mean()
    sigma = hrdf[col].std()
    minus1 = mu - sigma ; plus1 = mu + sigma
    minus2 = mu - (2*sigma) ; plus2 = mu + (2*sigma)
    minus3 = mu - (3*sigma) ; plus3 = mu + (3*sigma)
    x1 = hrdf[(hrdf[col] > minus1) & (hrdf[col] < plus1)]
    x2 = hrdf[(hrdf[col] > minus2) & (hrdf[col] < plus2)]
    x3 = hrdf[(hrdf[col] > minus3) & (hrdf[col] < plus3)]
    return(x1,x2,x3)
```
